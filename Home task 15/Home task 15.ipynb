### dataset

```python
import pandas as pd 

df = pd.read_csv("breast_cancer.csv")
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df[:20]
```

#### scale

```python
from sklearn.preprocessing import StandardScaler

df_features = df.drop(['diagnosis'], axis=1)
scaler  = StandardScaler().fit(df_features)
X_scaled = scaler.transform(df_features)
```

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
Z = pca.fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def diagnosis(x):
    if x =='M':
        return 1
    else:
        return 0

plt.figure(figsize=(10,8))
colors = np.array(['yellow','purple'])
df_diagnosis= df['diagnosis'].apply(diagnosis)
plt.scatter(Z[:,0], Z[:,1], c= colors[df_diagnosis],  s= 80, edgecolor = 'black', label='negative', alpha = 0.7)   
```
